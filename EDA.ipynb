```python
df.groupby("Region")["Sales"].sum()
```

```python
df.groupby("Region")["Sales"].sum().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Sales by Region")

plt.show()
```

```python
df.groupby("Sub-Category")["Profit"].sum().sort_values()
```

```python
df.groupby("Sub-Category")["Profit"].sum().sort_values().plot(
    kind="barh",
    figsize=(10,8)
)

plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()
```

```python
monthly_sales.plot(
    figsize=(12,5)
)

plt.title("Monthly Sales Trend")

plt.show()
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum()

top_customers.sort_values(ascending=False).head(10)
```

```python
top_customers.sort_values(ascending=False).head(10).plot(
    kind="bar"
)

plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum()

top_products.sort_values(ascending=False).head(10)
```

```python
top_products.sort_values(ascending=False).head(10).plot(
    kind="barh"
)

plt.show()
```
